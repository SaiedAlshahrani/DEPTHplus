# Quantifications of Bot Activity in All Wikipedia Editions:

```python
from time import sleep
from pathlib import Path
import warnings, pandas as pd

from wikistats2csv.helper import Helper
from wikistats2csv.contributing import Contributing

warnings.simplefilter("ignore", UserWarning)
warnings.simplefilter("ignore", FutureWarning)

wikis = Helper.get_Wikis_Codes()
```

## @ Quantifications of Bot-Created Articles in All Wikipedia Editions:

```python
%rm Bots/*--new-pages--bot-articles--all-years-exc-last-month--monthly.csv

counter = 1
botArticles = pd.DataFrame(columns=['Language', 'Wiki', 'botArticles'])
for k, v in wikis.items():
    filename = f'{v.lower()}--new-pages--bot-articles--all-years-exc-last-month--monthly.csv'
    filename = filename.replace(' ','-')
    print(f'[{counter}] {filename}')
    Contributing.new_pages(wiki=k, period='all-years-exc-last-month', filter='bot-articles', interval='monthly')
    counter += 1 ; sleep(3)
    
    bot_articles = pd.read_csv(f'Bots/{filename}')
    bot_articles = sum(bot_articles['total.name-bot']) + sum(bot_articles['total.group-bot'])
    botArticles = botArticles.append({'Language': v, 'Wiki': k, 'botArticles': bot_articles}, ignore_index=True)

    
botArticles.to_csv('botArticles.csv',index=False)
```

## Quantifications of Bot-Made Edits on Articles in All Wikipedia Editions:

```python
%rm Bots/*--edits--bot-edits--all-years-exc-last-month--monthly.csv

counter = 1
botEdits = pd.DataFrame(columns=['Language', 'Wiki', 'botEdits'])
for k, v in wikis.items():
    filename = f'{v.lower()}--edits--bot-edits--all-years-exc-last-month--monthly.csv'
    filename = filename.replace(' ','-')
    print(f'[{counter}] {filename}')
    Contributing.edits(wiki=k, period='all-years-exc-last-month', filter='bot-edits', interval='monthly')
    counter += 1 ; sleep(3)

    bot_edits = pd.read_csv(f'Bots/{filename}')
    bot_edits = sum(bot_edits['total.name-bot']) + sum(bot_edits['total.group-bot'])
    botEdits = botEdits.append({'Language': v, 'Wiki': k, 'botEdits': bot_edits}, ignore_index=True)
    
    
botEdits.to_csv('botEdits.csv',index=False)
```
